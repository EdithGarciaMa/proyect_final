```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
%matplotlib inline
```

```python
noche_estrellada = cv2.imread('./Images/Noche_Estrellada.jpg')
noche_estrellada_color = cv2.cvtColor(noche_estrellada, cv2.COLOR_BGR2RGB)
#noche_estrellada_gray = cv2.cvtColor(imGirasol_color, cv2.COLOR_RGB2GRAY)

plt.imshow(noche_estrellada_color, cmap='gray')
print(noche_estrellada_color.shape)
#plt.imshow(filtro_color)
```

```python
# source https://docs.opencv.org/master/d1/db7/tutorial_py_histogram_begins.html
color = ('r','g','b')
fig = plt.figure(figsize=(5, 3), dpi=100)
for i,col in enumerate(color):
    hist = cv2.calcHist([noche_estrellada_color],[i],None,[256],[0,256])
    plt.plot(hist,color = col)
    plt.xlim([0,256])
plt.legend(['r','g','b'])
plt.show()
plt.savefig('noche_Hist_Color.png')
fig.savefig('noche_Hist_Color.png',dpi=200)
```

```python
# Se importan las librerias necesarias de skimage
from skimage import data
from skimage import exposure
from skimage.exposure import match_histograms

# se cargan las imágenes 
# en este caso el histograma de image será modificado
# usando el histograma de reference.


image = prueba
reference = filtro_color

# Se llama a la función match_histograms
matched = match_histograms(image, reference, multichannel=True)

#Se muestran los resultados. 
# Se genera el contenedor de las figuras en un grid de 1x3 y se define el tamaño de la figura
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(8, 3),
                                    sharex=True, sharey=True)

# se eliminan los ejes de las gráficas
for aa in (ax1, ax2, ax3):
    aa.set_axis_off()

# se muestra y se agregan los títulos a cada una de las imágenes
ax1.imshow(image)
ax1.set_title('Source')
ax2.imshow(reference)
ax2.set_title('Reference')
ax3.imshow(matched)
ax3.set_title('Matched')

# se muestra la gráfica.
plt.tight_layout()
plt.show()
```
